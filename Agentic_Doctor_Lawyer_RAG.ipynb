```python
# CELL 2 - IMPORTS
import os, pandas as pd
from pathlib import Path
from datetime import datetime
from getpass import getpass
from langchain_groq import ChatGroq
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.prompts import ChatPromptTemplate
```

```python
# CELL 3 - GROQ SETUP
os.environ["GROQ_API_KEY"] = getpass("Enter Groq API Key: ")
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.2,
    groq_api_key=os.environ["GROQ_API_KEY"]
)
```

```python
# CELL 4 - UPLOAD PDFs
from google.colab import files
uploaded = files.upload()
```

```python
# CELL 5 - EMBEDDINGS
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# CELL 6 - BUILD DOCTOR DB
doctor_docs = PyPDFLoader("doctor_knowledge.pdf").load()
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
doctor_chunks = splitter.split_documents(doctor_docs)
doctor_db = FAISS.from_documents(doctor_chunks, embeddings)
doctor_db.save_local("doctor_db")
```

```python
# CELL 7 - BUILD LAWYER DB
lawyer_docs = PyPDFLoader("lawyer_knowledge.pdf").load()
lawyer_chunks = splitter.split_documents(lawyer_docs)
lawyer_db = FAISS.from_documents(lawyer_chunks, embeddings)
lawyer_db.save_local("lawyer_db")
```

```python
# CELL 8 - LOAD RETRIEVERS
doctor_db = FAISS.load_local("doctor_db", embeddings, allow_dangerous_deserialization=True)
lawyer_db = FAISS.load_local("lawyer_db", embeddings, allow_dangerous_deserialization=True)
doctor_retriever = doctor_db.as_retriever(search_kwargs={"k":4})
lawyer_retriever = lawyer_db.as_retriever(search_kwargs={"k":4})
```
